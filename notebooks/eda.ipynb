What is Individual Training Grants (ITGs)?

NYC program that pay for tuition, registration fees, testing fees, and books for in-demand trainings in certain occupations and at eligible training providers. Offer no-cost training programs for jobs in the industrial and construction fields

Why? Proactively address the skills gap and upskill New Yorkers.
